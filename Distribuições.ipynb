Importando Bibliotecas

```python
import random as r
import matplotlib.pyplot as fig
from scipy.stats import norm
import statistics as st
import numpy as np
```

Gerando as distribuições em listas

```python
aleat = []
gauss = []
gamma = []

for i in range(1000):
    x = r.uniform(-1, 1) #Entre 0 - 1
    g = r.gauss(0, 1) #Média, e desvio padrão
    z = r.gammavariate(1, 2) #Alpha -> Achatamento, Beta -> Decaimento
    aleat.append(x)
    gauss.append(g)
    gamma.append(z)
```

Distribuição de números randômicos de uma distribuição uniforme

```python
fig.subplot(211, title = "Uniforme")
fig.plot(aleat)
fig.subplot(212)
fig.hist(aleat, bins=10)
```

Distribuição de números randômicos de uma distribuição gaussiana

```python
fig.figure()
fig.subplot(211, title = "Gaussiana")
fig.plot(gauss)
fig.subplot(212)
fig.hist(gauss, bins=10)
```

Distribuição de números randômicos de uma distribuição gamma

```python
fig.figure()
fig.subplot(211)
fig.plot(gamma)
fig.subplot(212)
fig.hist(gamma, bins=10)
```

```python
fig.hist(gauss, bins=10, density = 1) #Normalizar para poder comparar
xmin, xmax = fig.xlim()
med = st.mean(gauss)
desv = st.pstdev(gauss)
eix = np.linspace(xmin, xmax, 100)
eiy = norm.pdf(eix, med, desv)
fig.plot(eix, eiy, color = 'black')
```

Gráfico da distribuição gaussiana cumulativa

```python
eiy = norm.cdf(eix, med, desv)
fig.plot(eiy)
```
