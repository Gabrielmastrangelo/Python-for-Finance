Importação das bibliotecas

```python
import xlrd
import numpy as np
import matplotlib.pyplot as fig
import matplotlib.mlab as m
```

Extraindo dados de uma planilha (na qual possui as cotações do IBOV), e adicionando a um vetor.

```python
wb = xlrd.open_workbook('ibov.xlsx')
p = wb.sheet_by_name('Planilha1')

lin = p.nrows
close = np.zeros(lin)

for i in range(lin):
    close[i] = p.cell(i, 4).value
```

Criando um vetor para armazenar a regressão linear

```python
x = np.linspace(1, lin, lin)

coef = np.polyfit(x, close, 1) #Porque é linear, 2 seria quadratica, 3 seria cubica

tend = coef[0]*x+coef[1] #Reta
```
